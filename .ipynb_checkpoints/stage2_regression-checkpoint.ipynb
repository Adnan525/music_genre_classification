## About Dataset
The dataset we are going to use for the task of pattern recognition and machine learning is the GTZAN music genre classification dataset. #### Feature Description
This dataset captures ten important sound features:

1. Chroma: Represents the twelve different pitch classes as features.
2. Root Mean Square Energy: The root mean square of the signal's amplitude.
3. Spectral Centroid: Represents the center of mass of the spectrum (indicating bright or dark sound).
4. Spectral Bandwidth: Measures the width of the spectral band.
5. Spectral Rolloff: The frequency below which a specified percentage of the total spectral energy lies.
6. Zero Crossing Rate: The number of times the audio waveform crosses the zero axis.
7. Harmony: Represents elements such as chords and their progression within each audio track.
8. Percept: Relates to the pitch of the sound.
9. Tempo: Pertains to rhythmic features.
10. MFCC (Mel-Frequency Cepstral Coefficients): Computed as a set of coefficients that capture spectral characteristics (20 in this dataset).

Each feature has two columns: mean and variance, with MFCC having 20 coefficients, so there are two for each coefficient. Increasing the number of MFCC coefficients reveals finer spectral details. You can add or remove columns as needed; typically, the range of 13-20 coefficients is used.

#### Principal Component Analysis
Principal Component Analysis (PCA) is a dimensionality reduction technique widely employed in data analysis and machine learning. Its objective is to simplify complex datasets by transforming them into a new set of uncorrelated variables known as principal components. These components are arranged in order of their ability to capture the maximum variance in the data. By retaining a subset of these components, PCA reduces the dimensionality of the dataset while preserving as much relevant information as possible. It is valuable for various purposes, including visualizing high-dimensional data, alleviating multicollinearity issues, and enhancing the performance of machine learning models by reducing computational complexity and noise.

PCA can also be leveraged to improve the visualization of data distributions. The code below will execute the PCA analysis and plot the results.

We are dividing the dataset to separate the labels from the features. The features are stored in the 'X' variable, while the string labels are stored in the 'y' variable.

```python
X = dataset.iloc[:,:-1] #features
y = dataset.iloc[:,-1] #labels
```

To prepare the dataset for PCA, we need to transform it to make it more uniform. As seen in the summary statistics above, each feature has its own spread, minimum, maximum, and variance, which makes it challenging to extract information from multidimensional data. To address this issue, we will scale the descriptive features using a Standard Scaler available in the sklearn.preprocessing package. This scaler function transforms each column in such a way that they have a mean of 0 and a standard deviation of 1.

For the classification task, it's also crucial to encode the labeled data with numerical values. This is performed below using the LabelEncoder() function from the same package. When a model classifies data, it provides numerical output. Therefore, having a string array to compare with numerical output can lead to errors.

The scaled features are shown below, and you can observe how all values are now spread across the range of -1 to 1.

```python
#Standard Scaler to transform X
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#LabelEncoder
label_encoder = LabelEncoder()

# Encode the class labels in y with numeric values
y_encoded = label_encoder.fit_transform(y)

#Display scaled features
X_df = pd.DataFrame(X_scaled, columns=X.columns)
X_df.head()
```

Now, we apply PCA transformation to our scaled feature data. Setting n_components to 2 means that PCA will reduce the dimensionality of the dataset by retaining the two most significant principal components. These two components capture the maximum variance in the data, providing a simplified representation of the dataset while preserving as much relevant information as possible.

```python
#PCA transformation
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

The code below provides the explained variance value for each of the principal components. A higher value indicates that the component has captured more information about the data's variation. For both components, the explained variance is around 0.2, indicating that each component accounts for only 20% of the variation in the data. This reveals that using PCA may result in a significant loss of information, regardless of dimensionality reduction. We will explore other techniques and methods to address this issue when finalizing the model later.

```python
#Calculating ecplained variance as part of PCA evaluation.
explained_variance = pca.explained_variance_ratio_
print("Explained Variance:", explained_variance)
```

The code below plots both the principal components for each class, and we can observe a significant overlap of data points from different genres. The PCA visualization does not provide a clear separation between classes, which may not be ideal for our purposes. As an alternative, we might consider using the original scaled dataset for our model. In later stages, we can explore feature selection techniques to reduce dimensionality and computational complexity.

```python
#Plot of principal components
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_encoded, cmap='viridis')
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Visualization")
plt.show()
```

## Data Modeling
#### Spliting Data
For model creation, we need to divide our data into two portions. One portion will be used to train the machine learning algorithm, while the other will be used to evaluate the model on unseen data. The split ratio can vary, but typically, the training portion is larger because the model requires sufficient data for training. In this case, we are going to use an 80-20 split, with 80% of the data for training and 20% for testing. We will use the 'X_scaled' variable to split the feature data, rather than 'X_pca,' as the results from PCA were not satisfactory. 

"train_test_split()" funtion from "sklearn.model_selection" package isused to perform the split.

```python
#Train-test split
test_size = 0.20
seed = 101 #for randomization of split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=test_size, random_state=seed)
```

#### Model Implementation
For this task, we are going to apply four different classification models to our data and evaluate their validation accuracy to select one for further fine-tuning in later stages. There are numerous variations and experiments that can still be explored, such as choosing a different dimensionality reduction method or utilizing the larger dataset available for the entire task. These aspects will be tested in subsequent stages.

The selected models to implement and their importance for solving this problem are described below.

##### Logistic Regression 
Logistic Regression is chosen for its simplicity and interpretability. It's suitable for binary classification tasks, making it an initial choice to establish a baseline performance. Its coefficients can provide insights into the influence of audio features on genre prediction. Logistic regression captures patterns from the data even if the relationships are non-linear as seen in our datset. 
Logistic Regression models the probability that a given music track belongs to a particular genre. It calculates a weighted sum of the input features and applies the logistic function to produce a probability between 0 and 1 or more. The model then assigns the track to the genre with the highest probability.
##### K-nearest Neighbor(KNN)
KNN classifies music tracks based on the majority class among their K nearest neighbors in the feature space. It calculates distances between the test instance and all training instances and selects the K closest neighbors. The model assigns the genre that is most frequent among these neighbors to the test instance. KNN is selected for its simplicity and ability to capture local patterns in the data. It can be effective when genres share similarities in audio feature space.
##### Support Vector Machines(SVMs)
SVMs aim to find the optimal hyperplane that best separates different music genres in a high-dimensional feature space. They do this by maximizing the margin between classes, and they can use various kernel functions to capture non-linear relationships in the data. SVMs are selected for their effectiveness in handling complex classification tasks. They can handle both linear and non-linear relationships in the data, making them versatile.
##### Random Forest
Random Forests consist of an ensemble of decision trees. Each tree is constructed using a bootstrapped subset of the dataset and a random subset of features. During classification, the model aggregates the predictions of individual trees to make a final prediction. Feature importance is determined based on how often features are used for splitting nodes across all trees. Random Forests are included as an ensemble method to reduce overfitting and enhance model robustness.The feature importance scores provided by Random Forests can reveal which audio features contribute most to genre classification.

The code chunk below defines all four models and saves them in a list called 'models.' The hyperparameters are randomly selected to perform a spot check on all the models and see if they start learning. Fine-tuning of hyperparameters is left for the next stage of this project, where we will make the same models learn more effectively.

```python
#List of models
models = []
models.append(('LR', LogisticRegression(solver='liblinear', multi_class='ovr'))) #Logistic Regression
models.append(('KNN', KNeighborsClassifier()))                                   #K-Nearest Neighbor
models.append(('SVM', SVC(gamma='auto')))                                        #Support Vector Machine
models.append(('RF', RandomForestClassifier(n_estimators=100, random_state=7)))  #Random Forest
```

#### Training and Cross Validation
Now it's time to train the models using the feature dataset. To evaluate how well the models have learned without using the test dataset for predictions, we employ cross-validation techniques. In cross-validation, the training dataset is divided into K subsets or folds. Each time, one or more of these folds are left out for validation, and the model is trained on the remaining data. Multiple combinations of training and validation folds are used to facilitate better learning.

The code below trains all the models one by one with K set to 5. This means that the data is split into 5 portions, and one portion is reserved for validation. The results display the validation score of each model, which represents the mean accuracy across all the training-validation combinations. In this case, accuracy is used as the validation score, indicating the percentage of correct predictions over all predictions on the validation set.

```python
# evaluate each model in turn
results = [] #store accuracy score for each model
names = [] #name of the corresponding model
for name, model in models:
	kfold = KFold(n_splits=5, random_state=seed, shuffle=True)  #Cross validation K = 5
	cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring='accuracy')
	results.append(cv_results)
	names.append(name)
	msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
	print(msg)
```

#### Model Comparision
The cross validation resuts look apealing, even without the hyper-perameter tuning cross-validation accuracy for all the models is close to 70% which  with SVM and random forest performing a better than KNn and logistic regression. Robustness to changing data and dealing with non-linearities and could be the reasons for better performance of SVM and random forest. However, there is alot to improve when tweeking the parameters. The code chunk below plots the model results using a bar plot where you can see the variation in accuracy scores for all the combinations of cross-validaion. 

The plot dipicts that the overall accuracy spread for random forest look appealing folowed by SVM because their accuracy score does not drop behind like in the case of logistic regression. Interpretation of the plot tells us that SVM and random forest model are more robust in learning patterns from this dataset. 

```python
# Compare Algorithms
fig = plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

## Performance
Below are four code chunks, one for each model. These code segments will calculate accuracy, plot a confusion matrix, and print a classification report for each model, one by one. In this phase, the models are being tested by predicting the genre class based on patterns in unseen data from X-test. The predictions are then compared to the actual class labels, and accuracy is calculated. The confusion matrix visually shows how many predictions per class were correct and which ones were predicted incorrectly. We can analyze the confusion matrix to identify which classes are most frequently misrepresented and explore the reasons behind these misclassifications. Finally, the classification report provides values for precision, recall, and F-score for each class, which are valuable metrics for evaluating the performance of a classification model.

```python
# Make predictions on validation dataset
#Logistic Regression
lr = LogisticRegression(solver='liblinear', multi_class='ovr')
lr.fit(X_train, y_train)
predictions = lr.predict(X_test)
print(accuracy_score(y_test, predictions))
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

```python
# Make predictions on validation dataset
# K-Nearest Neighbor
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
predictions = knn.predict(X_test)
print(accuracy_score(y_test, predictions))
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

```python
# Make predictions on validation dataset
# Support Vector Machine
svm = SVC(gamma='auto')
svm.fit(X_train, y_train)
predictions = svm.predict(X_test)
print(accuracy_score(y_test, predictions))
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

```python
# Make predictions on validation dataset
# Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=7)
rf.fit(X_train, y_train)
predictions = rf.predict(X_test)
print(accuracy_score(y_test, predictions))
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```
